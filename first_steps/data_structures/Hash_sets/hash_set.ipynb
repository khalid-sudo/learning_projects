#Problem 1: Array Intersection =>  Pinpoint the elements that appear in both of the given lists
#Naive approache

```python

list1 = [3,5,10,54,3,45,33,21,33,24]
list2 = [4,8,3,76,987,65,3,5]
intersection_list = []
start_time = time.time()
for num1 in list1:
    if num1 not in intersection_list:
        for num2 in list2:
            if num1==num2:
                    intersection_list.append(num1)
                    list2.remove(num1)
                    break
    else:
        continue
print(f"it took {time.time() - start_time}'s") #output: it took 5.9604644775390625e-06's
print(intersection_list)
```

#Problem 2: Non-Repeating Elements => determine all elements in a given list that appear only once, meaning they don't have any duplicates in the same list

```python
#Naive approache
list1 = [3,5,10,54,3,45,32,5,21,33,33]
start_time = time.time()
unique = []
duplicate = []
for num in range(len(list1)):
    for j in range(num+1,len(list1)):
        if list1[num] == list1[j]:
            duplicate.append(list1[num])
        elif j == len(list1) - 1 and list1[num]  not in duplicate:
                unique.append(list1[num])
                
print(sorted(unique))
print(f"it took {time.time() - start_time}'s") #output it took 1.621246337890625e-05's
```

```python
start_time = time.time()
seen, repeated = set(), set()
for num in list1:
    if num in seen:
        repeated.add(num)
    else:
        seen.add(num)
print(sorted(list(seen-repeated)))
print(f"it took {time.time() - start_time}'s") #it took 5.0067901611328125e-06's
```

#Problem 4: Unique String in the List consist on  identifying the first unique string from a list. 

```python
#naive_approach
string_sentence = "i really hope that we will learn together,  sharing is caring that's our emblem, our motto. Hope this example is just right to show case the difrence"

biglist = string_sentence.split(". ")
sublist = [subl for sub in biglist for subl in sub.split(", ")]
words = [s.lower() for sen in sublist for s in sen.split(" ")]

duplicated = []

for index in range(len(words)):
    if words[index] in duplicated:
        continue
    else:
        for index_j in range(index + 1,len(words)):
            if words[index] == words[index_j]:
                duplicated.append(words[index])
            
for word in words:
    if word not in duplicated:
        print(f"TADAAAA=>{word}")
    break
```

```python
#susing_set
seen, duplica = set(),set()

for word in words:
    if word in seen:
        duplica.add(word)
    seen.add(word)

for w in words:
    if w not in duplica:
        print(f"TADAAAA=>{w}")
        break
```

#Problem5: Anagram Pairs in Two Lists. identify all pairs of anagrams where each pair is made up of one word from the first list and one from the second list.

```python
#naive approach
list_1 = ["listen", "evil", "dusty", "night","heyoo","nice"]
list_2 = ["silent", "vile", "study", "thing", "helm", "ice"]
output = []
for word1 in list_1:
    for word2 in list_2:
        if tuple(sorted(word1))== tuple(sorted(word2)):
            output.append((word1, word2))
            break
print(output)
#this will work but as long as we have a small list Imagine the case where we'll have documents and we should do the comparision each time it will take an eternity
```

```python
#setway
#this section aims to narrow the scope of search we get the elemnt that has the same order of caracters 
output = []
list_1_set = set(tuple(sorted(w)) for w in list_1)
list_2_set = set(tuple(sorted(w)) for w in list_2)
elemnts_in_both_list = list_1_set & list_2_set

#now we don't have to compare with the whole list we have the elements in both sides
#in order to regain the initial shape 
the_anagram_elem_fromlist_1 = [word for word in list_1 if tuple(sorted(word)) in elemnts_in_both_list]
the_anagram_elem_fromlist_2 = [word for word in list_2 if tuple(sorted(word)) in elemnts_in_both_list]

for word1 in the_anagram_elem_fromlist_1:
    for word2 in the_anagram_elem_fromlist_2:
        if tuple(sorted(word1))== tuple(sorted(word2)):
            output.append((word1, word2))
            break
print(output) #  [('listen', 'silent'), ('evil', 'vile'), ('dusty', 'study'), ('night', 'thing')]
```

```python
#dictionary way
from collections import defaultdict
list_1 = ["listen", "evil", "dusty", "night","heyoo","nice"]
list_2 = ["silent", "vile", "study", "thing", "helm", "ice"]

mapping_1 = defaultdict(list)
for word in list_1:
    sorted_tuple = tuple(sorted(word))
    mapping_1[sorted_tuple].append(word)

mapping_2 = defaultdict(list)

for word in list_2:
    sorted_tuple = tuple(sorted(word))
    mapping_2[sorted_tuple].append(word)
    
common_tuples = set(mapping_1.keys()) & set(mapping_2.keys())


output = []

for anagram_tuple in common_tuples:

    for word1 in mapping_1[anagram_tuple]:

        for word2 in mapping_2[anagram_tuple]:


            output.append((word1, word2))
            
print(output)
```
