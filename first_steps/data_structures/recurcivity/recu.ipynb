Problem 1: fibonachi

```python
def fib(n, computed={0: 0, 1: 1}):

    if n not in computed:

        computed[n] = fib(n - 1, computed) + fib(n - 2, computed)

    return computed[n]
```

#Problem 2: Summing Array Elements

```python
def summ(arry, index=0):
    if index == len(arry):
        return 0
    else:
         arry[index] + sum(arry, index + 1)
```

#Problem 3: Calculating Factorial

```python
def factoriel(num: int) -> int:
    if num >= 0:
        if num <= 1:
            return 1 
        else:
            num * factoriel(num - 1)
```
